```python
import pandas as pd

# Extrage datele de temperatură la 2m
tmp_2m = dataset['TMP_2m']

# Convertirea la Pandas DataFrame pentru analiza ulterioară
df_tmp_2m = tmp_2m.to_dataframe().reset_index()


print(df_tmp_2m.head())
```

```python

print(df_tmp_2m.isnull().sum())


df_tmp_2m = df_tmp_2m.dropna()
```

```python
# Calcularea mediei temperaturii
mean_temperature = df_tmp_2m['TMP_2m'].mean()
print(f"Temperatura medie anuală: {mean_temperature}")


std_temperature = df_tmp_2m['TMP_2m'].std()
print(f"Deviația standard a temperaturii: {std_temperature}")

# Analiza valorilor maxime și minime
min_temp = df_tmp_2m['TMP_2m'].min()
max_temp = df_tmp_2m['TMP_2m'].max()
print(f"Temperatura minimă: {min_temp}, Temperatura maximă: {max_temp}")
```

```python
import matplotlib.pyplot as plt

# Plotează distribuția temperaturii
plt.hist(df_tmp_2m['TMP_2m'], bins=50, color='skyblue', edgecolor='black')
plt.title("Distribuția temperaturii")
plt.xlabel("Temperatura (K)")
plt.ylabel("Frecvența")
plt.show()

# Plotează temperaturile medii pe an (exemplu de analiză temporală)
df_tmp_2m['year'] = df_tmp_2m['time'].dt.year
mean_yearly_temp = df_tmp_2m.groupby('year')['TMP_2m'].mean()

mean_yearly_temp.plot(kind='line', color='orange', title="Temperatura medie anuală")
plt.xlabel("An")
plt.ylabel("Temperatura medie (K)")
plt.show()
```

```python
import numpy as np

# Extrage valorile de temperatură pe ani
years = df_tmp_2m['year'].unique()

# Calculul mediei temperaturii pentru fiecare an
yearly_means = []
for year in years:
    yearly_data = df_tmp_2m[df_tmp_2m['year'] == year]['TMP_2m']
    yearly_mean = np.mean(yearly_data)
    yearly_means.append(yearly_mean)

# Plotează rezultatele
plt.plot(years, yearly_means, label="Temperatura medie anuală")
plt.xlabel("An")
plt.ylabel("Temperatura medie (K)")
plt.title("Temperatura medie pe ani")
plt.legend()
plt.show()
```

```python
# Extrage datele pentru precipitații
apcp_sfc = dataset['APCP_sfc']
df_apcp = apcp_sfc.to_dataframe().reset_index()

# Combină temperatura și precipitațiile într-un singur DataFrame
df_combined = pd.merge(df_tmp_2m, df_apcp, on=['time', 'latitude', 'longitude'])

# Calculează corelația
correlation = df_combined[['TMP_2m', 'APCP_sfc']].corr()
print(correlation)
```

```python
# Verifică pentru valori lipsă
print(df_tmp_2m.isnull().sum())
```

```python
# Înlocuiește valorile lipsă cu media coloanei
df_tmp_2m['TMP_2m'].fillna(df_tmp_2m['TMP_2m'].mean(), inplace=True)
```

```python
# Extrage anul din coloana de timp
df_tmp_2m['year'] = df_tmp_2m['time'].dt.year

# Calculează media temperaturii pe an
mean_yearly_temp = df_tmp_2m.groupby('year')['TMP_2m'].mean()

# Plotează temperatura medie anuală
mean_yearly_temp.plot(kind='line', color='orange', title="Temperatura medie anuală")
plt.xlabel("An")
plt.ylabel("Temperatura medie (K)")
plt.show()
```

```python
# Extrage luna din coloana de timp
df_tmp_2m['month'] = df_tmp_2m['time'].dt.month

# Calculează media temperaturii pe lună
mean_monthly_temp = df_tmp_2m.groupby('month')['TMP_2m'].mean()

# Plotează temperatura medie lunară
mean_monthly_temp.plot(kind='bar', color='lightgreen', title="Temperatura medie lunară")
plt.xlabel("Lună")
plt.ylabel("Temperatura medie (K)")
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()
```

```python
# Selectează o regiune pe baza latitudinii și longitudinii
tmp_2m_region = df_tmp_2m[(df_tmp_2m['latitude'] >= 10) & (df_tmp_2m['latitude'] <= 20) &
                          (df_tmp_2m['longitude'] >= 65) & (df_tmp_2m['longitude'] <= 85)]

# Plotează temperaturile pentru această regiune
plt.plot(tmp_2m_region['time'], tmp_2m_region['TMP_2m'])
plt.title("Temperatura pentru regiunea selectată")
plt.xlabel("Timp")
plt.ylabel("Temperatura (K)")
plt.show()
```
